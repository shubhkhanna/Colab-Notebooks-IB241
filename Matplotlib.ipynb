```
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,5,11)
y = x**2
y
```

```
plt.plot(x,y,"red")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("First graph")

plt.show()
```

# Subplot

```
# plt.subplot(nrows,ncols,plot_number)
plt.subplot(1,2,1) #same row
plt.plot(x,y,'c--')

plt.subplot(1,2,2)
plt.plot(y,x,'g^-')

plt.show()
```

```
# plt.subplot(nrows,ncols,plot_number)
plt.subplot(2,1,1) #same column
plt.plot(x,y,'c--')

plt.subplot(2,1,2)
plt.plot(y,x,'g^-')

plt.show()
```

#Matplotlib Object Oriented Method

```
fig = plt.figure() 

axes = fig.add_axes([1.0,1.4,1.6,2.0])

axes.plot(x,y,'r--')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
```

# More than one axis 

```
fig = plt.figure() 

axes = fig.add_axes([0.1,0.4,0.8,0.8]) #left,bottom,width,height
axes1 = fig.add_axes([0.2,0.8,0.4,0.3])

axes.plot(x,y,'r--')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('Graph 1')

axes1.plot(y,x,'r--')
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes.set_title('Graph 2')
```

#More automatic axis manager

```
plt.subplot(211)
xx=np.linspace(-10,10,150)
cos=np.cos(xx)
sin=np.sin(xx)

plt.plot(xx,cos,color="red",linewidth=2.4)

plt.subplot(212)
plt.plot(xx,sin,color="green",linewidth=3.4)
```
