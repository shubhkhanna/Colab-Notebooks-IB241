```
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,5,11)
y = x**2
y
```

```
plt.plot(x,y,"red")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("First graph")

plt.show()
```

# Subplot

```
# plt.subplot(nrows,ncols,plot_number)
plt.subplot(1,2,1) #same row
plt.plot(x,y,'c--')

plt.subplot(1,2,2)
plt.plot(y,x,'g^-')

plt.show()
```

```
# plt.subplot(nrows,ncols,plot_number)
plt.subplot(2,1,1) #same column
plt.plot(x,y,'c--')

plt.subplot(2,1,2)
plt.plot(y,x,'g^-')

plt.show()
```
