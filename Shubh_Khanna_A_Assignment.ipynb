```
import pandas as pd
import numpy as np
```

```
carData = pd.read_csv('https://raw.githubusercontent.com/shubhkhanna/Colab-Notebooks-IB241/master/DataSet/Car_sales.csv')
carData
```

```
carData.isnull().sum()
```

```
carData.median()
```

```
carData['x_4_year'].fillna(carData['x_4_year'].median(),inplace=True)
carData['Price in thousands'].fillna(carData['Price in thousands'].median(),inplace=True)
carData['Engine size'].fillna(carData['Engine size'].median(),inplace=True)
carData['Horsepower'].fillna(carData['Horsepower'].median(),inplace=True)
carData['Wheelbase'].fillna(carData['Wheelbase'].median(),inplace=True)
carData['Width'].fillna(carData['Width'].median(),inplace=True)
carData['Length'].fillna(carData['Length'].median(),inplace=True)
carData['Curb weight'].fillna(carData['Curb weight'].median(),inplace=True)
carData['Fuel capacity'].fillna(carData['Fuel capacity'].median(),inplace=True)
carData['Fuel efficiency'].fillna(carData['Fuel efficiency'].median(),inplace=True)
carData['XX'].fillna(carData['XX'].median(),inplace=True)
```

```
carData.isnull().sum()
```

```
carData = carData.drop(['Unnamed: 15'],axis=1)
carData
```

```
#chnaging name of the column
carData.rename(columns={'x_4_year':'4_Year_Resale_Value'},inplace=True)
carData
```

```
newCarData = carData.iloc[:,[0,1,2,13]]
newCarData
```

```
df = newCarData.groupby('Sales in thousands').max()
df
```

```
lowSale = df[df['Fuel efficiency'] >= 20]
lowSale
```

```
highSale = lowSale[lowSale['Fuel efficiency'] <= 40]
highSale
```

```
highSale.tail(1)
```

```
carData = carData.iloc[:,[0,1,2,3,7]]
new = carData.groupby('Sales in thousands').min()
horse=new[new['Horsepower'] > 300]
horse.head(1)
```

# Placement Data

```
plaData = pd.read_csv('https://raw.githubusercontent.com/shubhkhanna/Colab-Notebooks-IB241/master/DataSet/Placement_Data.csv')
plaData
```

```
new = plaData[plaData['workex'] == 'Yes']
hs = new.groupby('salary').max()
hs.head(1)
```

```
new = plaData[plaData['workex'] == 'No']
deg=new[new['degree_t'] == 'Sci&Tech']
hs = deg.groupby('salary').min()
hs.head(1)
```

```
new = plaData[plaData['workex'] == 'No']
hs = new.groupby('salary').max()
ssc=hs[hs['ssc_p'] >= 66]
ssc.tail(1)
```

```
new = plaData[plaData['workex'] == 'No']
hs = new.groupby('salary').max()
deg=hs[hs['hsc_s']=='Commerce']
hdeg=deg[deg['degree_t']=='Sci&Tech']
hdeg
```
